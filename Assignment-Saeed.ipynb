```python
import pandas as pd


def NewUser(data_frame):

    username = input('Enter new username: ')
    password = input('Enter new password: ')
    special_question = input('Enter a special question: ')
    special_answer = input('Enter a special answer: ')
    
    new_user = {
        'Username' : username,
        'Password' : str(password),
        'Special Question' : str(special_question),
        'Special Answer' : str(special_answer)
    }

    new_user = pd.DataFrame([new_user])
    if username in data_frame['Username'].values:
        print(f'{username} account already exists')
    else:
        data_frame = pd.concat([data_frame, new_user], ignore_index=True)
        
        data_frame.to_csv('CSV_Module.csv', index=False)
    
    return data_frame
    
def SignIn(data_frame):
    username = input('Enter your username: ')
    password = input('Enter your password: ')
    print(f'Answer this question for verification:')
    if username in data_frame['Username'].values:
        special_question = data_frame.loc[data_frame['Username'] == username, 'Special Question'].iloc[0]
        print(special_question)
        special_answer = input('Enter its special answer: ')

    if data_frame.empty:
        print('There is no any account\nTry to create an Account!')
        print('Create an account here >')
        NewUser(data_frame)
    else:
        if (data_frame[(data_frame['Username'] == username) & (data_frame['Password'] == password) & (data_frame['Special Answer'] == special_answer)]).shape[0] > 0:
            print(f'Welcome {username}')
        else:
            print('Incorrect Username/Password/Answer!\nTry again!')
            
def ResetPassword(data_frame):
    username = input('Enter your username: ')
    password = input('Enter new password: ')
    print(f'Answer this question for verification:')
    if username in data_frame['Username'].values:
        special_question = data_frame.loc[data_frame['Username'] == username, 'Special Question'].iloc[0]
        print(special_question)
        Special_answer = input('Enter its special answer: ')

        special_answer = data_frame.loc[data_frame['Username'] == username, 'Special Answer'].iloc[0]
        if Special_answer == special_answer:
            data_frame.loc[data_frame['Username'] == username, 'Password'] = password
            data_frame.to_csv('CSV_Module.csv', index=False)
            print('Password has been successfully updated XD')
        else:
            print(f'Incorrect Answer!')
    else:
        print(f'No {username} account!')
    
    

if __name__ == '__main__':

    DF_file = pd.read_csv('CSV_Module.csv')
    
    print('-- 1. New User --')
    print('-- 2. Sign In --')
    print('-- 3. Forget Password ? --')
    choice = int(input('Enter your choice: '))

    if choice == 1:
        NewUser(DF_file)
    elif choice == 2:
        SignIn(DF_file)
    elif choice == 3:
        ResetPassword(DF_file)
    else:
        print('Invalid choice!')
            
    
```
